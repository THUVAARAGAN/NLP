```python
def pattern_match(pattern,text_discribtion):
  matches = re.findall(pattern,text_discribtion)
  return matches[0]
```

```python
print(pattern_match('\(age.*\n(.*)',text))
print(pattern_match('[a-z0-9A-Z_]*@[a-z0-9A-Z_]*\.com | [a-z0-9A-Z_]*@[_a-z0-9A-Z]*\.lk' , chat5))
```

```python
def persional_info(Text):
  name = pattern_match('Born(.*)',Text)
  age = pattern_match('age (\d+)',Text)
  place = pattern_match('\(age.*\n(.*)',Text)
  date = pattern_match('Born.*\n(.*)\(age',Text)

  return {
      'name': name.strip(),
      'age': int(age),
      'place':place.strip(),
      'date': date.strip()
  }
```

```python
persional_info(text)
```
